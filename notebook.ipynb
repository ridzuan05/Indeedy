Using Indeedy
=============
This notebook uses `indeedy` (https://github.com/petebachant/Indeedy) to count the number of job postings for various keywords on Indeed.com.

### Example: Programming languages for scientists in California

```python
language = ["Python",
            "C",
            "Java",
            "MATLAB",
            "LabVIEW",
            "FORTRAN"]

njobs, salaries = indeedy.compare_jobs_title(language, constant="big data", location="Kuala Lumpur")
indeedy.dual_bar_graph(language, njobs, salaries)
```
